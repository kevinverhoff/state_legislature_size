```python
import pandas as pd
```

```python
df_leg = pd.read_csv('data\state_legislatures_size.csv')
df_leg = df_leg.drop(['population_per_representative_2000','population_per_senator_2000','population_of_state_2000'],axis=1)
df_leg = df_leg.drop(0,axis=0)
df_leg['States'] = df_leg['States'].map(lambda x: x.rstrip(' (SAH,*)'))
df_leg = df_leg.rename(columns={'States':'State'})

population = []
for index, row in df_leg.iterrows():
    p = float(row['population_of_state_2010'].replace(',',''))
    population.append(p)
df_leg['population'] = population
df_leg = df_leg.drop(['population_of_state_2010'],axis=1)

senate_seats = []
for index, row in df_leg.iterrows():
    p = float(row['number_of_seats_senators'].replace(',',''))
    senate_seats.append(p)
df_leg['senate_seats'] = senate_seats
df_leg = df_leg.drop(['number_of_seats_senators'],axis=1)

rep_seats = []
for index, row in df_leg.iterrows():
    p = float(row['number_of_seats_representatives'].replace(',',''))
    rep_seats.append(p)
df_leg['rep_seats'] = rep_seats
df_leg = df_leg.drop(['number_of_seats_representatives'],axis=1)

pop_per_senator = []
for index, row in df_leg.iterrows():
    p = float(row['population_per_senator_2010'].replace(',',''))
    pop_per_senator.append(p)
df_leg['pop_per_senator'] = pop_per_senator
df_leg = df_leg.drop(['population_per_senator_2010'],axis=1)

pop_per_rep = []
for index, row in df_leg.iterrows():
    p = float(str(row['population_per_representative_2010']).replace(',',''))
    pop_per_rep.append(p)
df_leg['pop_per_rep'] = pop_per_rep
df_leg = df_leg.drop(['population_per_representative_2010'],axis=1)

df_leg['total_seats'] = df_leg['rep_seats'] + df_leg['senate_seats']
df_leg['pop_per_legislator'] = round(df_leg['population']/df_leg['total_seats'],0)
df_leg
```

```python
df_legparty = pd.read_csv('data\state_legislature_breakdown_2019.csv')
dff_legparty = df_legparty[['STATE','dem_percent','rep_percent']]
dff_legparty = dff_legparty.rename(columns={'STATE':'State','dem_percent':'leg_dem','rep_percent':'leg_rep'})
dff_legparty
```

```python
df_pres = pd.read_csv('data\state_by_state_2016_presidential.csv')
dff_pres = df_pres[['State','votePctDem','votePctRep']]
dff_pres = dff_pres.rename(columns={'votePctDem':'pres_dem','votePctRep':'pres_rep'})
dff_pres
```

```python
df_comp = dff_pres.merge(dff_legparty,on='State')
df_comp
```

```python
df = df_leg.merge(df_comp,on='State')
df['total_diff'] = (abs((df['pres_dem']/100)-df['leg_dem']) + abs((df['pres_rep']/100)-df['leg_rep'])/2)
df
```

```python
df_ab = pd.read_csv('data\states_abbreviations.csv')
df = df.merge(df_ab,on='State')
df
```

```python
df_geo = pd.read_csv('data\geographic_centers_states.csv')
df = df.merge(df_geo,on='State')
df
```

```python
df.to_csv('data\legislatures_clean.csv')
```
